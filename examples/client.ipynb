```python
from grafa.entity.search import extract_concepts

user_query = "que es el bias?"
concepts = (await extract_concepts(user_query, c.llm)).entities
```

```python
queries = [user_query] + concepts
import asyncio

per_query_limit = 5
search_mode = "allowed"

# Create a list of search tasks for each query
search_tasks = []
for query in queries:
    # Create a task for each similarity search
    task = c.similarity_search(
        query=query,
        node_types=None,  # Search across all user-defined node types
        limit=per_query_limit,  # Limit results per query
        search_mode=search_mode,  # Use both semantic and text-based search
    )
    search_tasks.append(task)

# Execute all search tasks concurrently
search_results = await asyncio.gather(*search_tasks)
```
